```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier

from src.kmeans import distance

data = load_iris()
```

```python
X = data['data']
y = data['target']
```

```python
model = KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X, y)
```

```python
X_test = data['data'][:10]
X_test
```

```python
y_test = data['target'][:10]
y_pred = model.predict(X_test)
model.score(X_test, y_test)
```

```python
(X - X_test[:,np.newaxis]).shape
```

```python
distances = np.sqrt((X-X_test[:,np.newaxis])**2).sum(axis=2)
```

```python
nearest_neighbors = np.argsort(distances, axis=1)[:, :5]
nearest_neighbors
```

```python
top_k_labels = y[nearest_neighbors]
top_k_labels
```

```python
from scipy.stats import mode
mode(top_k_labels,  axis=1)[0]
```
